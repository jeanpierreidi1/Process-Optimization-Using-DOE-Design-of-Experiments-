```python

import pandas as pd
import os
import matplotlib.pyplot as plt

#Load data

df = pd.read_csv('C:/Users/jeanp/Downloads/doe_project/doe_process_optimization_raw_data.csv')

# Display data
print("Dataset:")
print(df.head())
```

```python
# Analysis by factors

print("\nAverage Quality by Temperature:")
print(df.groupby("Temperature")["Quality"].mean())

print("\nAverage Quality by Time:")
print(df.groupby("Time")["Quality"].mean())

print("\nAverage Quality by Material:")
print(df.groupby("Material")["Quality"].mean())
```

```python
# Best condition
best = df.loc[df["Quality"].idxmax()]
print("\nBest Condition:")
print(best)
```

```python
df.groupby("Temperature")["Quality"].mean().plot(kind="bar")
plt.title("Quality by Temperature")
plt.show()
```
